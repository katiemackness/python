```python
import pandas as pd
import os
iris = pd.read_csv("iris.csv")

# Aplica las condiciones
iris_sepal = iris[(iris["sepal.width"].astype(float) >= 3.5) & (iris["sepal.length"].astype(float) > 5.0)]

# Cree una tabla con la media, desviacion estandar y la mediana
iris_sepal.pivot_table(values = "petal.width", index = 'variety', aggfunc = ["mean", "median", "std"])
```

```python
# Calcule el numero de registros
iris_sepal['variety'].value_counts()
```

```python
# Asegurese de que la tabla es una copia completa
iris_sepal = iris_sepal.copy()

# Agregue una nueva columna del estandar
iris_sepal.loc[:, 'petal.width.st'] = (iris_sepal['petal.width'] - iris_sepal['petal.width'].mean()) / iris_sepal['petal.width'].std()
print(iris_sepal)
```

```python
# Cambia el nombre de columna
iris_sepal.rename(columns = {"variety": "Variedad", "petal.width.st": "Ancho Est. PÃ©talo"}, inplace=True)
iris_sepal
```

```python
# Exporte el resultado
iris_sepal.to_csv("Iris2.csv")
```
