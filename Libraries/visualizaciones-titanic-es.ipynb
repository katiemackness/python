```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('titanic.csv')
```

```python
# El Histograma de frecuencias para la variable edad 
fig, ax = plt.subplots()
ax.hist(df["Age"])
ax.set_xlabel("Edad")
ax.set_ylabel("# de pasajeros")
plt.show()
```

```python
# Descargar figura
fig.savefig("edad_pasajeros_hist.jpg")
```

```python
# El diagrama de caja para la variable edad
fig, ax = plt.subplots()
ax.boxplot([df["Age"]])
ax.set_xticklabels(["Pasajeros"])
ax.set_ylabel("Edad de pasajero")
plt.show()
```

```python
# Descargar figura
fig.savefig("edad_pasajeros_box.jpg")
```

```python
# Un gr√°fico que muestre el n√∫mero de pasajeros de cada clase (Pclass)
df = pd.read_csv('titanic.csv', index_col="Pclass")
```

```python
# Visualizaci√≥n de pasajeros por clase
fig, ax = plt.subplots()
ax.hist(df.index, bins = [1, 2, 3, 4], )
ax.set_title('Numero de pasajeros por clase') 
ax.set_xticks([1,2,3])
ax.set_xlabel("Clase")
ax.set_ylabel("# de pasajeros")
plt.show()
```

```python
# Descargar figura
fig.savefig("pasajeros_clase_hist.jpg")
```
